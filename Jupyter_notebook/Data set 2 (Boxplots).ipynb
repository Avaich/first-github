```python
import seaborn as sns
import pandas as pd
import numpy as np

#importing data file
chilla2 = pd.read_csv("Chilla_data2_for_plots.csv")
chilla2.head()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
#Canvas (Baloon board)
sns.set_theme(style="ticks", color_codes=True)
#Load dataset
sns.boxenplot(x="Qualification_completed", y="Age (years)-Float/Int", hue="Gender", data=chilla2)
```

## Renaming x axis, y axis and title of the graph

```python
sns.boxplot(x="Qualification_completed", y="Age (years)-Float/Int", hue="Gender", data=chilla2)
plt.xlabel("Qualification Completed"),
plt.ylabel("Age(years)"),
plt.title("Box plot of Qualification vs Age in Male/Female")
```

## Changing size and making text bold

```python
sns.boxplot(x="Qualification_completed", y="Age (years)-Float/Int", hue="Gender", data=chilla2)
plt.xlabel("Qualification Completed", size=14),
plt.ylabel("Age(years)", size=12),
plt.title("Box plot of Qualification vs Age in Male/Female", size=16, weight="bold")
```

## Changing color of each hue color

```python
sns.boxplot(x="Qualification_completed", y="Age (years)-Float/Int", hue="Gender", data=chilla2, palette={"Male": "0.4", "Female": "0.8"})
plt.xlabel("Qualification Completed", size=14),
plt.ylabel("Age(years)", size=12),
plt.title("Box plot of Qualification vs Age in Male/Female", size=16, weight="bold")
```
