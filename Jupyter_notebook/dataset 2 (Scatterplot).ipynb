```python
import seaborn as sns
import pandas as pd
import numpy as np

#importing data file
chilla2 = pd.read_csv("Chilla_data2_for_plots.csv")
chilla2.head()
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
sns.relplot(
    data=chilla2,
    x="Purpose_for_chilla", y="Age (years)-Float/Int",
    hue="Where do you live?", size="Where do you live?")  
```

## Changing color

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
sns.relplot(
    data=chilla2,
    x="Purpose_for_chilla", y="Age (years)-Float/Int",
    hue="Where do you live?", size="Where do you live?", palette={"Urbun": "b", "Rural": "0.85", "International": "c"}) 
```

## Changing Size

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
sns.relplot(
    data=chilla2,
    x="Purpose_for_chilla", y="Age (years)-Float/Int",
    hue="Where do you live?", size="Where do you live?", palette={"Urbun": "b", "Rural": "0.85", "International": "c"})
plt.xlabel("Purpose of Chilla", size=15),
plt.ylabel("Age(years)", size=15),
plt.title("Purpose of Chilla WRT to Age and Location", size=15, weight="bold") 
```
